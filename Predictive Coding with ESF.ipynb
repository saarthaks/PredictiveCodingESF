## Persistent Activity
In this setting, we assume the network implements an integrator circuit, where we provide a brief 2 dimensional driving input $c(t)$ and design our network to "store" the integrated value. These dynamics are given by:
$$ \dot{x}(t) = c(t)$$

```python
T = 1
dt = 5e-5
ts = np.arange(0,T,dt)
S = len(ts)

J = 2
c = np.zeros((J,S))
x = np.zeros_like(c)
lam_x = 0
c[0,:500] = 30
c[1,:500] = 20
for t in range(1,S):
    x[:,t] = x[:,t-1] + dt*c[:,t-1]

plt.figure(figsize=(8,5))
plt.plot(ts, x[0,:], label='dim 0')
plt.plot(ts, x[1,:], label='dim 1')
plt.legend()
```

## Variability of Spike Trains
When realized with random uncorrelated noise on the neuron's membrane potential or threshold, the ESF is designed to exhibit high trial-to-trial variability in its resulting spike train. A common metric for this variability is the Fano factor, empirically computed for a given neuron using the spike count in trial $j$, $N^j$, with the following equation:
$$FF = \frac{Var[N^j]}{E[N^j]}$$

For Poisson spiking models, originally proposed to reflect empirically observed variability, one would expect $FF = 1$ for each neuron. Values of $FF>1$ suggest higher variability than Poisson spiking, and values of $FF<1$ suggest lower variability than Poisson spiking.

```python
N = 100
lam_r = 8
G_scale = np.sqrt(2) * 0.01
sig_noise = 1e-5
trial_count = 10
np.random.seed(30)

xh_ff, spktm_ff, o_ff, ff = simulate_ff(dt, S, N, J, G_scale, lam_x, lam_r, x, c, sig_noise, trials=trial_count)
rmse = compute_rmse(x, xh_ff[0])
```

```python
plt.figure(figsize=(8,5))
titl = 'RMSE = %.6f, Mean FF = %.4f' % (rmse, ff)
plt.title(titl)
for j in range(x.shape[0]):
    lbl = 'est dim %d' % j
    plt.plot(ts, x[j,:])
    plt.plot(ts, xh_ff[0][j,:], label=lbl)
plt.legend()

lmax = 1000
ni = 10
R = np.zeros((2*lmax-1,))
o1 = o_ff[0][ni,lmax:-lmax+1]
o2 = o_ff[1][ni,:]
frate = np.mean([np.sum(o1),np.sum(o2)])/T
for l in range(len(R)):
    R[l] = np.dot(o1, o2[l:-(2*lmax-1-l)])/frate

plt.figure(figsize=(8,5))
titl = 'Autocorrelation Function Normalized by Firing Rate'
plt.title(titl)
plt.plot(1e3*dt*np.arange(-lmax,lmax-1), R)
plt.xlim([-45,45])
plt.xlabel('Time (ms)');
```

## Autoencoder
As our overarching system, we'll task our network with approximating an "autoencoder" which low-pass filters an arbitrary time-varying 2D input $c(t)$ for $t\in [0,T]$ through the following dynamics:
$$\dot{x}(t) = -\lambda_x x(t) + c(t)$$

```python
np.random.seed(5)
T = 1
dt = 5e-5
ts = np.arange(0,T,dt)
S = len(ts)

J = 2
lam_x = 10
c = generate_signal(J, 0, T, dt)
x = filter_signal(c, dt, lam_x)
plt.figure(figsize=(8,5))
plt.plot(ts, x[0,:], label='dim 0')
plt.plot(ts, x[1,:], label='dim 1')
plt.legend()
```

## Tight E/I Balance

Feedback from inhibitory neuron population closely follows the feedforward and recurrent excitatory inputs, allowing neurons to precisely track an error projection. We induce mismatch between these populations by adding uncorrelated Gaussian noise to the synaptic weights which modulate the inhibitory currents within the excitatory population, and can observe the resulting root-mean-squared error and the variance in the net E-I current.

```python
N = 100
lam_r = 10
lam_s = 100
G_scale = np.sqrt(2) * 0.01
Th = 0.5 * G_scale**2
mismatch_noise = 1e-8/(1.25**np.arange(9))
np.random.seed(1)

xh_bal, Ie_bal, Ii_bal, rmse_bal, var_bal = simulate_balance(dt, S, N, J, G_scale, lam_x, lam_r, lam_s, 
                                                             x, c, mismatch_noise, trials=1)
```

```python
plt.figure(figsize=(12,6))
plt.loglog(mismatch_noise/Th, rmse_bal.values(), '-*k')
plt.ylabel('RMSE')
plt.xlabel('Variance of E/I Recurrent Weights (normalized by threshold)')
plt.figure(figsize=(12,6))
plt.loglog(mismatch_noise/Th, var_bal.values(), '-*k')
plt.ylabel('Variance of E-I')
plt.xlabel('Variance of Mismatch in E/I Recurrent Weights (normalized by threshold)')
```

```python
@interact(mismatch_var=list(mismatch_noise))
def plot_noise(mismatch_var=list(mismatch_noise)):
    ni = 1
    ylim = [-0.003,0.003]
    plot_reconstruction(ts, x, xh_bal[mismatch_var], rmse_bal[mismatch_var])
    plot_balance(ni, ts, Ie_bal[mismatch_var]+Ii_bal[mismatch_var], var_bal[mismatch_var], ylim)
```

## Robustness to Noise

The claim is that these networks are robust to forms of random uncorrelated noise. We model this by injecting white Gaussian synaptic current with $\mu_I=0$ and variance $\sigma_I^2$. We look at the firing patterns and root-mean-squared error.

```python
N = 40
lam_r = 3
G_scale = np.sqrt(2) * 0.01
Th = 0.5 * G_scale**2
sig_noise = 1 / (2**(np.arange(13,22)))
np.random.seed(25)

xh_noise, spktm_noise, rmse_noise = simulate_noise(dt, S, N, J, G_scale, lam_x, lam_r, x, c, sig_noise)
```

```python
plt.figure(figsize=(12,6))
plt.semilogx(sig_noise, rmse_noise.values(), '-*k')
plt.ylabel('RMSE')
plt.xlabel('Variance of Somatic Current Noise')
```

```python
@interact(noise_var=list(sig_noise))
def plot_noise(noise_var=list(sig_noise)):
    plot_reconstruction(ts, x, xh_noise[noise_var], rmse_noise[noise_var])
    plot_raster(N, T, spktm_noise[noise_var])
```

## Ablation

What happens when neurons die? Does the network fall apart? For each inactive neuron, we zero out its feedforward and recurrent weights and observe the reconstruction. This is done over a range of values $\rho$, defined as the proportion of inactive neurons in the network.

```python
N = 100
lam_r = 3
G_scale = np.sqrt(2) * 0.01
Th = 0.5 * G_scale**2
rho = np.arange(0,60,5)/N
np.random.seed(10)

xh_ablate, spktm_ablate, rmse_ablate = simulate_ablation(dt, S, N, J, G_scale, lam_x, lam_r, x, c, rho)
```

```python
plt.figure(figsize=(8,5))
plt.plot(rho, rmse_ablate.values(), '-*k')
plt.ylabel('RMSE')
plt.xlabel('Inactive Network Proportion')
```

```python
@interact(p=list(rho))
def plot_ablate(p=list(rho)):
    plot_reconstruction(ts, x, xh_ablate[p], rmse_ablate[p])
    plot_raster(N, T, spktm_ablate[p], rho=p)
```

## Uniform Synaptic Transmission Delays
Instantaneous spike propagation is an unrealistic idealization, and more realistic transmission dynamics can be modeled by low-pass filtering the synaptic input currents. Self-resets are still assumed instantaneous.

```python
N = 40
lam_r = 20
G_scale = np.sqrt(2) * 0.01
Th = 0.5 * G_scale**2
delay = 1000*np.arange(1,23,2)
np.random.seed(50)

xh_delays, spktm_delays, rmse_delays = simulate_delays(dt, ts, S, N, J, G_scale, lam_x, lam_r, x, c, delay)
```

```python
plt.figure(figsize=(8,5))
plt.semilogx(1000/delay, rmse_delays.values(), '-*k')
plt.ylabel('RMSE')
plt.xlabel('Transmission Delay (ms)')
```

```python
@interact(d=list(delay))
def plot_delays(d=list(delay)):
    plot_reconstruction(ts, x, xh_delays[d], rmse_delays[d])
    plot_raster(N, T, spktm_delays[d])
```
