```python
import numpy as np
from matplotlib import pyplot as plt
import random
%matplotlib inline

def compute_rmse(x,x_hat):
    return np.sqrt(np.mean((x-x_hat)**2))

def generate_signal(I,tmin,tmax,dt,fmax=4,amp=1):
    trange = np.arange(0,tmax-tmin,dt)
    tlen = len(trange)
    sig = np.zeros((I,tlen))
    num_signals = 10
    for dim in range(I):
        for j in range(num_signals):
            p = fmax*2*np.pi*(2*random.random()-1)
            s = amp*(2*random.random()-1)
            sig[dim,:] = sig[dim,:] + s*np.sin(p*trange)
        sig[dim,:] = sig[dim,:] - sig[dim,0]
    
    return sig
```

```python
T = 1
dt = 1e-5
ts = np.arange(0,T,dt)
S = len(ts)

Ne = 2
Ni = 8

d = 2

c = generate_signal(d, 0, T, dt, fmax=3, amp=2)
l = 1
x = np.zeros_like(c)
for t in range(1,S):
    x[:,t] = (1-l*dt)*x[:,t-1] + dt*c[:,t-1]

plt.figure()
plt.plot(ts, x[0,:])
plt.plot(ts, x[1,:])
```

```python
G_scale = 0.01
Ne = 20
Ni = 400

G1 = G_scale * np.random.randn(d,Ne)
G1 = G_scale * (G1/np.linalg.norm(G1,axis=0)[np.newaxis,:])
G_mag = np.sum(G1**2,0).T
ThE = 0.5*G_mag

G2 = np.random.randn(Ne,Ni)
G2 = G_scale * (G2/np.linalg.norm(G2,axis=0)[np.newaxis,:])
G_mag = np.sum(G2**2,0).T
ThI = 0.5*G_mag

F = G1.T
Gpos = G1.copy(); Gpos[Gpos<=0] = 0
Gneg = -G1.copy(); Gneg[Gneg<=0] = 0
W_ee = Gpos.T@Gpos + Gneg.T@Gneg
W_ei = (Gpos.T@Gneg + Gneg.T@Gpos) @ G2
W_ie = G2.T
W_ii = G2.T@G2

noise_var = 1e-6
noise_E = noise_var*np.random.randn(Ne,S)
noise_I = noise_var*np.random.randn(Ni,S)
```

```python
lr = 2
h = np.exp(-lr*ts)

Ve = np.zeros((Ne,S))
Vi = np.zeros((Ni,S))
re = np.zeros((Ne,S))
ri = np.zeros((Ni,S))
oe = np.zeros((Ne,S))
oi = np.zeros((Ni,S))
e = np.zeros_like(x)
xh = np.zeros_like(x)
er = np.zeros_like(re)
reh = np.zeros_like(re)

for t in range(1,S):
    re_last = re[:,t-1]; oe_last = oe[:,t-1]
    ri_last = ri[:,t-1]; oi_last = oi[:,t-1]
    Ve_last = Ve[:,t-1]
    Vi_last = Vi[:,t-1]

    dVe = F@c[:,t-1] + l*W_ee@re_last - l*W_ei@ri_last - W_ee@oe_last + W_ei@oi_last + noise_E[:,t-1]
    Ve[:,t] = Ve_last + dVe*dt

    dVi = W_ie@oe_last - W_ii@oi_last + noise_I[:,t-1]
    Vi[:,t] = Vi_last + dVi*dt

    ne = np.argmax(Ve[:,t]-ThE)
    if np.abs(Ve[ne,t]) >= ThE[ne]:
        oe[ne,t] += 1/dt
        re[ne,t:] += h[:(S-t)]

    ni = np.argmax(Vi[:,t]-ThI)
    if np.abs(Vi[ni,t]) >= ThI[ni]:
        oi[ni,t] += 1/dt
        ri[ni,t:] += h[:(S-t)]

    xh[:,t] = G1@re[:,t]
    reh[:,t] = G2@ri[:,t]
        
        
#     e[:,t] = x[:,t]-xh[:,t]
#     Ve[:,t] = G1.T@e[:,t]

#     ne = np.argmax(Ve[:,t]-ThE)
#     while np.abs(Ve[ne,t]) >= ThE[ne]:
#         oe[ne,t] += 1/dt
#         re[ne,t:] += h[:(S-t)]
        
#         er[:,t] = re[:,t]-reh[:,t]
#         Vi[:,t] = G2.T@er[:,t]
        
#         ni = np.argmax(Vi[:,t]-ThI)
#         while np.abs(Vi[ni,t]) >= ThI[ni]:
#             oi[ni,t] += 1/dt
#             ri[ni,t:] += h[:(S-t)]
#             reh[:,t] = G2@ri[:,t]
#             er[:,t] = re[:,t]-reh[:,t]
#             Vi[:,t] = G2.T@er[:,t]
#             ni = np.argmax(Vi[:,t]-ThI)
        
#         xh[:,t] = G1@reh[:,t]
#         e[:,t] = x[:,t]-xh[:,t]
#         Ve[:,t] = G1.T@e[:,t]
#         ne = np.argmax(Ve[:,t]-ThE)
    
#     xh[:,t] = G1@reh[:,t]
```
